```python
def addRelativeIndices(path, key1, key2, indexCol, groupIndex):
    df = pd.read_csv(path)
    if('Unnamed: 0' in df.columns):
        df = df.drop(columns = ['Unnamed: 0'])
    df['unique_key'] = df[key1].astype(str) + "/" + df[key2].astype(str)
    df = df.sort_values([key1, key2])
    df[indexCol] = df.groupby(groupIndex).cumcount()+1
    df = df.rename(columns = {'Index': 'sample_index'})
    return df
```

```python
addRelativeIndices("./data/tidy/tidyLandmarks_AT_new.csv", "min_alpha", "min_theta", "landmark_index", "Index")
```
